**DESCRIPTION:**
         Decision tree algorithm belongs to the family of supervised learning algorithms. Unlike other supervised learning algorithms decision tree can be used to solve regression and classification problems. The goal of decision tree is to create training model that can predict class(single or multi) or value by learning simple decision rules from training data. Decision tree form a flow chart like structure thatâ€™s why they are very easy to interpret and understand. It is one of the few ML algorithm where its very easy to visualize and analyze the internal working of algorithm.
    

Just like flowchart, decision 
tree contains different types of nodes and branches. Every decision node represent the test on feature and based on the test result it will either form another branch or the leaf node. Every branch represents the decision rule and leaf node represent the final outcome.

**Importing the required libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

**Loading the iris dataset**

```python
iris = load_iris()
X=iris.data[:,:] 
y=iris.target
```

```python
data=pd.DataFrame(iris['data'],columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data['species']=iris['target']
data['species']=data['species'].apply(lambda x: iris['target_names'][x])

data.head()
```

**Exploratory Data Analysis**

```python
data.shape
```

```python
data.describe()
```

**Data Visualization**

```python
sns.pairplot(data=data,hue='species')
```

```python
sns.FacetGrid(data,hue='species',height=4,aspect=1).map(plt.scatter,'Sepal Length','Sepal Width').add_legend()
plt.show()

# Scatter plot of data based on Petal Length and Width features
sns.FacetGrid(data,hue='species',height =4,aspect=1).map(plt.scatter,'Petal length','Petal Width').add_legend()
plt.show()
```

Model Training

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1) 
tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(X_train,y_train)
print('Decision Tree Classifer Created')
```

```python
y_pred = tree_classifier.predict(X_test)
```

**Comparing Actual and Predicted Classification**

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
df
```

**Visualizing trained model**

```python
df.plot(kind='bar',figsize=(10,8))
```

```python
from sklearn import tree
fig = plt.figure(figsize=(20,10))
_ = tree.plot_tree(tree_classifier, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

```python
sample = [[1,2,2,3]]
```

```python
tree_classifier.predict(sample)
```

**Model Accuracy**

```python
import sklearn.metrics as sm
print("Accuracy:",sm.accuracy_score(y_test, y_pred))
```
